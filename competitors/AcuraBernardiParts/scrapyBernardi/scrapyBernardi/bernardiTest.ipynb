```python
db = pymysql.connect(host='localhost',user='root',password='123456',db='spiders')
cursor = db.cursor()
sql = 'select * from acurabernardiparts;'
cursor.execute(sql)
data = cursor.fetchall()
db.close()

data
```

```python
UrlData = pd.DataFrame(data, columns=['Brand','Url','Competitor','PartNumber','PartsUrl'])
UrlData
```

```python
url = 'https://acura.bernardiparts.com/search.aspx?keyword=74120-S6M-506'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36 Edg/114.0.1823.43',
}
```

```python
PartsUrl = UrlData['PartsUrl']
dit = []
for i in PartsUrl:
    try:
        response = requests.get(i,headers=headers)
        html = etree.HTML(response.text)
        yourprice = html.xpath('//tr[@class="yourPrice"]//span[@class="SitePrice"]/text()')
        PartNumber = html.xpath('//*[@id="ctl00_MainContentHolder_lblSku"]/text()')
        dit.append({'PartNumber':PartNumber, 'YourPrice': yourprice})
    except:
        print('falied')
```
